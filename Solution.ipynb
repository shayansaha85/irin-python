```python
n=int(input("Number of elements:"))
lst=[]
for i in range(n):
    x=int(input("Enter Element"))
    lst.append(x)
count=0
count1=0
for i in range(len(lst)):
    if lst[i]%2==0:
      count+=1
    else:
        count1+=1
print("no of even nm:",count)
print("no of odd nm:",count1)
    
```
